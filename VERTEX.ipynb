```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Load data
# Open the target data file from local storage and parse the file’s comma-separated rows, converting them to a DataFrame object.
# Clean up formatting inconsistencies in the file headers, trimming off spaces and fixing quote marks.
# Output an ordered list of clean column names to the terminal console to confirm file loading.
# -----------------------------
data = pd.read_csv('withola13.csv')
data.columns = data.columns.str.replace('"', '').str.strip()
print("Columns in data:", data.columns.tolist())

# -----------------------------
# Sampling parameters
# Define the sampling frequency (must match the timing loop configured in ESP32 sensor firmware).
# Calculate the continuous stepping interval between adjacent digital samples (dt = 0.002 seconds).
# Extract the absolute log time from the time column and convert to seconds.
# Measure the total number of rows (discrete samples) within the dataset, setting the array size boundaries.
# Set up index lists to map the raw data names to clear labels.
# -----------------------------
fs = 500.0  # Hz
dt = 1.0 / fs
time = data['time'].values / 1000.0  # convert ms to seconds
N = len(data)

axes = ['x1','y1','z1','x2','y2','z2']
titles = ["MPU1 (response) X-axis", "MPU1 (response) Y-axis", "MPU1 (response) Z-axis",
          "MPU2 (control) X-axis", "MPU2 (control) Y-axis", "MPU2 (control) Z-axis"]

# -----------------------------
# 1. Time-series plots
# Allocate a grid of 6 independent subplots arranged across 2 rows and 3 columns.
# Flatten the 2D grid matrix into a simple 1D linear array of 6 elements (allowing to loop through subplots with a single variable)
# Iterate through each sensor axis while maintaining a counter.
# Plot the continuous time vector on the horizontal axis against the raw acceleration values from the data frame on the vertical axis. 
# Automatically adjust subplot margins to prevent overlapping labels or titles.
# Render the chart window to the user’s screen.
# -----------------------------
fig, axs = plt.subplots(2, 3, figsize=(18, 8))
axs = axs.flatten()
for i, col in enumerate(axes):
    axs[i].plot(time, data[col].values, linewidth=0.8)
    axs[i].set_title(titles[i])
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel("Acceleration (g)")
plt.tight_layout()
plt.show()

# -----------------------------
# 2. Time-series plots with moving RMS
# Multiply the target window duration by the sampling rate to determine the required window size in samples.
# Compute the symmetric half-width window size used to center the window around each data point.
# Zero-center the data by subtracting the current means from every sample (removing biases).
# Initialize a vector of zeros matching the length of the dataset to store values.
# Loop through every index inside the dataset (handling edge effects).
# Implement the discrete RMS formula on the active window segment.
# Overlay the bright red moving RMS profile directly onto the zero-centered acceleration trace.
# -----------------------------
window_sec = 1.0
window_size = int(window_sec * fs)
half_window = window_size // 2

fig, axs = plt.subplots(2, 3, figsize=(18, 8))
axs = axs.flatten()
for i, col in enumerate(axes):
    y = data[col].values - np.mean(data[col].values)
    rms_moving = np.zeros(N)
    for j in range(N):
        start = max(0, j - half_window)
        end = min(N, j + half_window + 1)
        rms_moving[j] = np.sqrt(np.mean(y[start:end]**2))
    axs[i].plot(time, y, linewidth=0.8, label='Acceleration')
    axs[i].plot(time, rms_moving, color='red', linewidth=1.2, label='Moving RMS')
    axs[i].set_title(titles[i])
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel("Acceleration (g)")
    axs[i].legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 3. RMS of each axis (entire signal)
# Calculate a single global RMS value for each axis across the entire dataset
# -----------------------------
rms_time_domain = {col: np.sqrt(np.mean(data[col].values**2)) for col in axes}
for col, val in rms_time_domain.items():
    print(f"Axis {col}: Time-domain RMS = {val:.4f} g")

# -----------------------------
# 4. FFT snapshots (33%,66%,100%)
# Define three target areas of time to evaluate how the signal’s frequency content changes over time.
# Force the axis array into a grid structure, matching each sensor axis against the three time snapshots.
# Locate the specific row index corresponding to the target percentage point in time.
# Extract a local 1-second data block centered around that point in time.
# Run the FFT algorithm on the windowed segment with NumPy.
# Generate the correct frequency labels for the horizontal axis based on the window length and step size.
# Normalize the complex numbers into physical units of power spectrum (g2)
# -----------------------------
snapshot_percents = [0.33,0.66,1]
fig, axs = plt.subplots(len(axes), len(snapshot_percents), figsize=(18,12))
axs = axs.reshape(len(axes), len(snapshot_percents))
for i, col in enumerate(axes):
    y = data[col].values - np.mean(data[col].values)
    for j, perc in enumerate(snapshot_percents):
        center = int(N * perc)
        start = max(0, center - half_window)
        end = min(N, center + half_window + 1)
        window = y[start:end]
        W = np.fft.fft(window)
        freqs = np.fft.fftfreq(len(window), dt)[:len(window)//2]
        power = (2.0 / len(window)) * np.abs(W[:len(window)//2])**2
        axs[i,j].plot(freqs, power)
        axs[i,j].set_title(f"{col} - snapshot {int(perc*100)}%")
        axs[i,j].set_xlabel("Frequency (Hz)")
        axs[i,j].set_ylabel("Power (g²)")
        axs[i,j].grid(True)
plt.tight_layout()
plt.show()

# -----------------------------
# 5. RMS of maximum FFT
# Compute the frequency spectrum across the entire data trace for a given axis.
# Find the single highest power spike in the spectrum and take its square root.
# -----------------------------
rms_max_fft = {}
for col in axes:
    y = data[col].values - np.mean(data[col].values)
    W = np.fft.fft(y)
    power = (2.0 / N) * np.abs(W)**2
    rms_max_fft[col] = np.sqrt(np.max(power))
print("RMS of max FFT:", rms_max_fft)

# -----------------------------
# 6. RMS vs dominant frequency (scatter plots)
# Group the sensors into matching pairs for comparison.
# Perform a sliding analysis across the entire dataset sample by sample.
# Find the index of the highest peak in the local power spectrum
# Look up the corresponding value in the frequency array (dominant frequency during 1-second window)
# Create a scatter lot mapping dominant frequency on the horizontal axis against moving RMS on the vertical axis.
# -----------------------------
accel_axes = [('x1','x2'),('y1','y2'),('z1','z2')]
fig, axs = plt.subplots(2,3,figsize=(18,8))
axs = axs.flatten()
for idx, (a1,a2) in enumerate(accel_axes):
    for sensor, ax in zip([a1,a2],[axs[idx],axs[idx+3]]):
        y = data[sensor].values - np.mean(data[sensor].values)
        rms_list = []
        dom_freq_list = []
        for i in range(N):
            start = max(0, i - half_window)
            end = min(N, i + half_window + 1)
            window = y[start:end]
            rms = np.sqrt(np.mean(window**2))
            W = np.fft.fft(window)
            freqs_win = np.fft.fftfreq(len(window), dt)[:len(window)//2]
            power = (2.0/len(window)) * np.abs(W[:len(window)//2])**2
            dom_freq = freqs_win[np.argmax(power)]
            rms_list.append(rms)
            dom_freq_list.append(dom_freq)
        ax.scatter(dom_freq_list, rms_list, s=5)
        ax.set_xlabel('Frequency (Hz)')
        ax.set_ylabel('RMS (g)')
        ax.set_title(f'RMS vs Freq: {sensor}')
        ax.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------
# 7. RMS and dominant frequency over time (12 subplots)
# Pack the calculated time histories of moving RMS and dominant frequency.
# Allocate a wide 12-panel dashboard.
# Render the moving RMS energy histories across the top row of 6 plots in red.
# Render the tracked dominant frequency histories across the bottom row of 6 plots in blue.
# -----------------------------
results = {}
for col in axes:
    y = data[col].values - np.mean(data[col].values)
    rms_list = np.zeros(N)
    dom_freq_list = np.zeros(N)
    for i in range(N):
        start = max(0,i-half_window)
        end = min(N,i+half_window+1)
        window = y[start:end]
        rms_list[i] = np.sqrt(np.mean(window**2))
        W = np.fft.fft(window)
        freqs_win = np.fft.fftfreq(len(window), dt)[:len(window)//2]
        power = (2.0/len(window)) * np.abs(W[:len(window)//2])**2
        dom_freq_list[i] = freqs_win[np.argmax(power)]
    results[col] = {"rms": rms_list, "dom_freq": dom_freq_list}

fig, axs = plt.subplots(2,6,figsize=(24,8))
axs = axs.flatten()
for i,col in enumerate(axes):
    axs[i].plot(time, results[col]["rms"], color='red', linewidth=0.8)
    axs[i].set_title(f"RMS: {col}")
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel("RMS (g)")
    axs[i].grid(True)
for i,col in enumerate(axes):
    axs[i+6].plot(time, results[col]["dom_freq"], color='blue', linewidth=0.8)
    axs[i+6].set_title(f"Dom Freq: {col}")
    axs[i+6].set_xlabel("Time (s)")
    axs[i+6].set_ylabel("Frequency (Hz)")
    axs[i+6].grid(True)
plt.tight_layout()
plt.show()
```
